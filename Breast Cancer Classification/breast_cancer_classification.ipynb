# Breast Cancer Classification
Given a dataset of properties of individuals with breast tumor. Predict if the tumor is malignant or not. The used model is based on a neural network.

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(42)
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
```

## Data preparation

```python
# load breast cancer data from csv
df = pd.read_csv("data.csv")
df
```

```python
# let's check info
df.info()
```

```python
# let's check if there's any missing value
df.isnull().any()
```

```python
# let's take a look at basic statistics
df.describe()
```

```python
# let's create a function for checking if there's any outlier
def check_outliers(column_start, column_end):
    df[df.columns[column_start:column_end]].plot(
        kind='box', 
        subplots=True, 
        sharey=False, 
        figsize=(12, 6)
    )
    plt.subplots_adjust(wspace=0.5) 
    plt.show()
```

```python
# Let's check diagnosis
df["diagnosis"].value_counts()
```

```python
# we won't need id column
df.drop(columns = ["id"], axis = 1, inplace = True)

# also let's label diagnosis with malignant = 1 and benign = 0
df = pd.get_dummies(df, drop_first = True)
```

Data has no missing values. Malignant-bening ratio is not significantly inbalanced. It may contain some outliers, but they might hold valueable information, so no further processing is required. We can split data to features and target, and then start building the model.

## Building the model

```python
# split features and target
# features
X = df.drop(columns = "diagnosis_M", axis = 1)
# target
Y = df["diagnosis_M"]
```

```python
# train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
# Scale data with standardscaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# setting up the layers of the neural network (input, hidden and ouput)
model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)), 
                          keras.layers.Dense(20, activation="relu"), 
                          keras.layers.Dense(2, activation="sigmoid")])
```

```python
# compiling the neural network
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
# training the neural network
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
# plot accuracy
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("model accuracy")
plt.legend(["training accuracy", "validation accuracy"])
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.show()
```

```python
# plot loss
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.legend(["training loss", "validation loss"])
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

## Evaluate the model

```python
# make a prediction of the probability of getting malignant or benign tumor from each row in test data
Y_pred = model.predict(X_test)
# converting probability to actual labels
# in each row the index of the higher probability of the two outputs become the label)
Y_pred = np.argmax(Y_pred, axis=1)
```

```python
# get metrics of model working on test data
loss, accuracy = model.evaluate(X_test, Y_test)
print("accuracy: ", accuracy)
print("precision: ", precision_score(Y_test, Y_pred , average="macro"))
print("recall: ", recall_score(Y_test, Y_pred , average="macro"))
print("f1 score: ", f1_score(Y_test, Y_pred , average="macro"))
```

```python
# display confusion matrix
cm = confusion_matrix(Y_test, Y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

The model scored high on all relevant metrics. I can safely draw the conclusion: building the neural network was a success!
